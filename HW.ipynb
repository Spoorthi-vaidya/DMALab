```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

train_data = pd.read_csv("train_users_2.csv")
test_data = pd.read_csv("test_users.csv")
df_all = pd.concat((train_data, test_data), ignore_index = True)
```

```python
import seaborn as sns
%pylab inline

df_all.signup_app.value_counts(dropna=False).plot(kind='bar', color='red', rot=0)
plt.xlabel('Signup_App')
sns.despine()
```

```python
df_all.signup_method.value_counts(dropna=False).plot(kind='bar', color='blue', rot=0)
plt.xlabel('Signup_Method')
sns.despine()
```

```python
df_all.first_device_type.value_counts(dropna=False).plot(kind='bar', color='green', rot=0)
plt.xlabel('Device_Type')
sns.despine()
```

```python
df_all.language.value_counts(dropna=False).plot(kind='bar', color='purple', rot=0)
plt.xlabel('Language')
sns.despine()
```
